Objetivo: Validar a fronteira de decisão de um neurônio artificial e analisar como o viés (bias) e os pesos (weights) afetam a separação de classes.

## 2. Definição dos Parâmetros do Modelo (Neurônio)
Nesta etapa, será importado o Numpy para manipulação matemática e o Matplotlib para visualização.

```python
import numpy as np #Possui a função de Validar a fronteira de decisão de um neurônio artificial.
import matplotlib.pyplot as plt

np.random.seed(42) #Fixa a semente para garantir que qualquer pessoa que for revisar o script visualize o mesmo gráfico.
```

##2. Definição dos Parâmetros do Modelo (Neurônio)
Aqui será definido os coeficientes da reta. Matematicamente, a e b são os pesos ($w_1, w_2$) e c é o viés ($b$).




```python
# Parâmetros para a reta de decisão
a = -1  # Peso (weight) w1
b = 4   # Peso (weight) w2
c = 0.4 # Viés (Bias) - Controla o deslocamento da reta

# Se alterar'c', a reta sobe ou desce sem mudar o ângulo.
# Se alterar 'a' ou 'b', mudará a inclinação (rotação) da fronteira.
```

##3. Função de Plotagem e Fronteira de Decisão
A função abaixo converte a equação geral da reta  $ax + by + c = 0$ para a forma funcional $y = f(x)$ para que o gráfico seja gerado.

```python
def plotline(a, b, c):
    # O intervalo do eixo X
    x = np.linspace(-2, 4, 50)
    # Isolando 'y': y = (-a*x - c) / b
    y = (-a*x - c) / b

    plt.axvline(0, color='k', linewidth=1, linestyle='--') # Eixo Y
    plt.axhline(0, color='k', linewidth=1, linestyle='--') # Eixo X
    plt.plot(x, y, label='Fronteira de Decisão', color='orange')
    plt.grid(True, alpha=0.3)
    plt.legend()

plotline(a, b, c)
```

## 4. Teste de Classificação (Pontos de Dados)
Logo abaixo é inserido os pontos no espaço para validar se a "lógica do neurônio" classifica corretamente baseado no sinal do resultado.

```python
# Pontos de teste (x1, x2)
p1 = (2, 0.4)   # Ponto azul
p2 = (1, 0.6)   # Ponto vermelho
p3 = (3, -0.4)  # Ponto verde

plotline(a,b,c) #Aqui define a fronteira de decisão no gráfico

# Plotagem dos pontos
plt.scatter(p1[0], p1[1], color='b', s=100, label='Classe A')
plt.scatter(p2[0], p2[1], color='r', s=100, label='Classe B')
plt.scatter(p3[0], p3[1], color='g', s=100, label='Classe C')

# Cálculo de ativação que retorna o valor para a distância do ponto à reta.
# Se positivo = Classe 1 | se negativo = Classe 0
ret1 = a*p1[0] + b*p1[1] + c
print(f"Ponto 1 (Azul): {ret1:.2f} -> {'Acima' if ret1 > 0 else 'Abaixo'} da reta")

ret2 = a*p2[0] + b*p2[1] + c
print(f"Ponto 2 (Vermelho): {ret2:.2f} -> {'Acima' if ret2 > 0 else 'Abaixo'} da reta")

ret3 = a*p3[0] + b*p3[1] + c
print(f"Ponto 2 (Verde): {ret3:.2f} -> {'Acima' if ret3 > 0 else 'Abaixo'} da reta")
```

##5. Conclusão do Estudo de Caso

O modelo realizou com sucesso a separação dos pontos utilizando a equação linear. Este cálculo manual é exatamente o que a função $nn.Linear$ faz internamente. Em um ambiente de produção, não é definido os pesos manualmente, ao invés disso é utilizado o *Gradiente Descendente* para que a rede aprenda esses valores de $a, b, c$ através dos dados.
